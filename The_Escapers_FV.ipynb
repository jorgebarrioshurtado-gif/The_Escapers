```python
import time

world_map = {
    "R0": {
        "name": "The Snake Pit",
        "tagline": "LOG: You are in a pit of vipers. It smells like damp earth. Use the ladder to escape.",
        "objects": {
            "book": {"type": "MAN", "info": "LOG: Manual says: Use 'git cauldron' to repair spells."},
            "terrarium": {"type": "CACHE", "item": "wood", "status": 1},
            "cauldron": {"type": "LOCK", "req": "unreadable_page", "status": 0, "unlock_msg": "LOG: Page is now readable!"},
            "ladder": {"type": "DOOR", "req": "wood", "target": "R1", "status": 0}
        }
    }
}

# --- PLAYER STATE ---
state = {
    "current_room": "R0",
    "inventory": ["unreadable_page"], # Starting item
    "game_over": 0
}

# --- GLOBAL LOGIC ---
def show_room():
    room = world_map[state["current_room"]]
    print(f"\n--- {room['name']} ---")
    print(room["tagline"])
    print(f"Visible: {list(room['objects'].keys())}")

def interact(target):
    room = world_map[state["current_room"]]
    
    if target not in room["objects"]:
        print("LOG: You can't do that here.")
        return

    obj = room["objects"][target]
    # Logic for different types
    if obj["type"] == "MAN":
        print(obj["info"])

    elif obj["type"] == "CACHE":
        if obj["status"] == 1:
            state["inventory"].append(obj["item"])
            obj["status"] = 0 # Empty the cache
            print(f"LOG: You took {obj['item']}.")
        else:
            print("LOG: It's empty.")

    elif obj["type"] == "LOCK":
        if obj["req"] in state["inventory"]:
            obj["status"] = 1
            print(obj["unlock_msg"])
        else:
            print(f"LOG: You need {obj['req']} for this.")

    elif obj["type"] == "DOOR":
        if obj["req"] in state["inventory"]:
            state["current_room"] = obj["target"]
            state["game_over"] = 1
            print ("""Suddenly… you hear a click.
            A door unlocks.
            You step through… into the next challenge.""")
        else:
            print(f"LOG: The {target} is broken. Need {obj['req']}.")

# --- MAIN LOOP ---
def play():
    while state["game_over"] == 0:
        show_room()
        cmd = input("\nAction?: ").strip().lower()
        # if cmd == "quit": break
        interact(cmd)

def room_1():
    """
    Play room 1.
    """
    help_key = 0
    print("""As you enter the next room, you hear someone calling for help.
 It sounds like Miguel… but his voice is distant.""")
    print("""You realize you can’t help him yet — you need guidance first.""")
    print("""You enter Adi’s room.
 Adi looks at you and says nothing… but the challenge is clear.""")
    print("You must understand how to ask for help properly.")
    print("""Instead of immediately asking the teacher, you must follow the correct process:
 first Google (3), then Slack (2), and only then ask Adi (1).""")

    while help_key != 1:
        step1 = input("What do you do first? Use the number in parentheses () next to the options ")
        step2 = input("What do you do second? Use the number in parentheses () next to the options ")
        step3 = input("What do you do last? Use the number in parentheses () next to the options ")
        if step1 == "3" and step2 != "2":
            print("You started well, but the order is not complete.")
        elif step1 == "3" and step2 == "2" and step3 != "1":
            print("Almost correct, but Adi should be last.")
        elif step1 == "3" and step2 == "2" and step3 == "1":
            print("Correct!")
            help_key = 1
        else:
            print("Wrong order. Try again.")

    print("Only when you follow this order correctly… everything starts to make sense.")
    print("You receive the Flow Control Key.")
    print("Now you are ready… to help Miguel.")

room_2 = {"wall":"You can see some words written on it. You are able to read ‘While loops’.",
"fridge":"Don’t worry Miguel, the fridge is not over 0º celsius.",
"washing machine":"Miguel stop! Don’t put the tomato sauce inside the washing machine",
"laptop":"What kind of python loop can go on forever?"}

def action():
    """
    this function is for moving around the room. it returns the action in lowercase.
    """
    print ("You see the laptop, a fridge, a washing machine and a wall")
    action = input("What do you want to search?")
    action_lowercase = action.lower()
    return action_lowercase

def password():
    """
    this function is to enter the password. if it is the correct one, it changes the door_lock variable.
    """
    password = input("Enter the password: ")
    password_lowercase = password.lower()
    if password_lowercase == "while loops":
        print ("""Congrats! You enter the correct password: ‘while loops’. 
        The system accepts it. 
        The door unlocks. 
        Miguel is free.""")
        door_lock = 1
        return door_lock
    else:
        print ("Incorrect password." )
        door_lock = 0
        return door_lock

def room_miguel():
    """
    Starts the room 2 and keep playing until you find the password
    """
    
    print ("""You rush into the next room and immediately see Miguel locked inside a prison. 
    He is trying to explain something, but you can’t hear him clearly.
    Around you, you see strange objects — a fridge, a washing machine, a wall… and a laptop asking for a password.""")
    door_lock = 0
    while door_lock != 1:
        action_lowercase = action ()
        if action_lowercase == "laptop":
            print ("You go to the", action_lowercase)
            print (room_2.get(action_lowercase))
            door_lock = password()
        elif action_lowercase in room_2.keys():
        	print ("You go to the", action_lowercase)
        	print (room_2.get(action_lowercase))
        else: 
            print ("Invalid action. Don't waste your time! Miguel isn't having a good time.")
    print("But before you can celebrate… another door opens.")

def room_3():
    """
    Play room 3.
    """
    total_mistakes = 0
    q1 = 0
    q2 = 0
    q3 = 0
    
    #reactions
    positive_sound = ["Ding-ding-ding!", "Ting-a-ling", "Chime-in", "Beep-beep", "Crowd applause", "Crowd scream"]
    negative_sound = ["Bzzzzt!", "Err-errr!", "Eh-err!", "Wanh-wanh", "Wunh-wunh", "Eeegh!", "BzzzzztBomp"]

    print ("""You enter Juliette’s room. This room feels different — calm, but intense.
    Juliette is sitting in front of a messy dataset, surrounded by screens filled with charts and numbers.
    She looks at you and says: ‘I’m close… but something is missing. I need your help to finish this.’
    The data is chaotic — missing values, inconsistent formats, confusing patterns.
    You realize this is the final challenge.
    She warns you: ‘You can only make three mistakes’""")
    ##################Q QUESTION 1 #####################
    while q1 != 1:
        mistakes(total_mistakes)
        if mistakes(total_mistakes) == 1:
            break
        print("Which are the two kinds of numerical data?")
        time.sleep(1)
        print("    " + "a) continuous and discrete")
        time.sleep(1)
        print("    " + "b) ordinal and continuous")
        time.sleep(1)
        print("    " + "c) nominal and ordinal")
        time.sleep(1)
        answer_1 = input("Please, select the correct letter a, b or c:")
        time.sleep(1)
        if answer_1 == "a":
            print(positive_sound [4] + ". Correct answer!")
            q1 = 1
        else:
            total_mistakes += 1
            print(negative_sound [5] + ". The answer is incorrect, +1 mistake!")
            print("You have ", total_mistakes, " mistakes")

    ##################Q QUESTION 2 #####################
    while q2 != 1:
        mistakes(total_mistakes)
        if mistakes(total_mistakes) == 1:
            break
        print("What kind of data is the blood type of a person?")
        time.sleep(1)
        print("    " + "a) numerical")
        time.sleep(1)
        print("    " + "b) booleans")
        time.sleep(1)
        print("    " + "c) categorical")
        time.sleep(1)
        answer_1 = input("please, select the correct letter a, b or c:")
        time.sleep(1)
        if answer_1 == "c":
            print(positive_sound [5] + ". Correct answer!")
            print("Every correct answer brings you closer to the truth.")
            q2 = 1
        else:
            total_mistakes += 1
            print(negative_sound [0] + ". The answer is incorrect, +1 mistake!")
            print("You have ", total_mistakes, " mistakes")
            
    ##################Q QUESTION 3 #####################
    while q3 != 1:
        mistakes(total_mistakes)
        if mistakes(total_mistakes) == 1:
            break
        print("How many times can you miss class unjustifiably?")
        time.sleep(1)
        print("    " + "a) 6 times")
        time.sleep(1)
        print("    " + "b) 3 times")
        time.sleep(1)
        print("    " + "c) 1 time")
        time.sleep(1)
        answer_1 = input("Please, select the correct letter a, b or c:")
        time.sleep(1)
        if answer_1 == "b":
            print(positive_sound [0] + ". Correct answer!")
            q3 = 1
        else:
            total_mistakes += 1
            print(negative_sound [2] + ". The answer is incorrect, +1 mistake!")
            print("You have ", total_mistakes, " mistakes!")
    if total_mistakes < 3:
        print("")
        print ("CONGRATULATIONS, YOU CAN GRADUATE NOW")
        print("")
        print("""And then… everything clicks.
    Juliette smiles: ‘You’re ready.’
    A final door slowly opens…""")
        print("")
        print("""You step into the final room.
 Your work is presented. There is silence… 
 Then suddenly: ‘You passed.’
 The walls disappear. The bootcamp fades away.
 Out of nowhere… a unicorn appears.
 Without hesitation, you jump on the unicorn and ride into freedom.
 The wind hits your face as you leave the bootcamp behind.
 A new chapter begins — your amazing, partly remote data career awaits.""") 
    else:
        print("")
        print ("Oh no!!! You lose all your opportunities, you feel dizzy and uneasy.")
        print ("GAME OVER.")
        print ("Start the game again to try to escape from Ironhack")  

def mistakes(mistakes):
    if mistakes == 3:
        return 1

def main():
    print("""You wake up and find yourself trapped inside an Ironhack Data Analytics bootcamp.
 At first, it feels like a normal day… until you realize something is wrong.
 The doors are locked. Your laptop is open. And challenges are waiting for you.
 The only way to escape is to complete every stage of the bootcamp.
 Each room represents a different part of your learning journey.
 Only by solving everything… you will graduate — and escape.""")
    print("")
    print("ROOM 0")
    print("")
    play()
    print("")
    print("ROOM 1")
    print("")
    room_1()
    print("")
    print("ROOM 2")
    print("")
    room_miguel()
    print("")
    print("ROOM 3")
    print("")
    room_3()
```
